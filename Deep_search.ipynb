```python
class Stack():
  def __init__(self):
    self.stack = []
    self.top = -1
  def isEmpty(self):
    if self.top == -1:
      return True
    else:
      return False
  def push(self, e):
    self.stack.append(e)
    self.top += 1
  def peek(self):
    if self.top == -1:
      print("항목이 없습니다.")
      return False
    return self.stack[self.top]
  def pop(self):
    if self.top == -1:
      print("항목이 없습니다.")
      return False
    self.top -= 1
    return self.stack.pop(self.top+1)
  def size(self):
    return self.top+1
  def clear(self):
    self.stack = []
    self.top = -1
  def display(self):
    for i in range(self.top + 1):
      print(i+1, "번째 나올 항목",  self.stack[self.top - i])
```

```python
class Search(Stack):
  def __init__(self):
    super().__init__()
  def startSearch(self, map):
    for x in range(len(map)):
      for y in range(len(map[0])):
        if map[x][y] == 'e':
          return [x,y]
  def deepSearch(self, map):
    currentPos = self.startSearch(map)
    already = [currentPos]
    while True:
      up = 1
      down = -1
      left = -1
      right = 1
      if currentPos[0] == 0:
        up = 0
      elif currentPos[0] == 5:
        down = 0
      if currentPos[1] == 0:
        left = 0
      elif currentPos[1] == 5:
        right = 0
      for wid in [left, right]:
        if map[currentPos[0] + wid][currentPos[1]] == '0':
          if [currentPos[0] + wid, currentPos[1]] not in already:
            self.push([currentPos[0] + wid, currentPos[1]])
            already.append([currentPos[0] + wid, currentPos[1]])
        elif map[currentPos[0] + wid][currentPos[1]] == 'x':
          return [currentPos[0] + wid, currentPos[1]]
      for hig in [down, up]:
        if map[currentPos[0]][currentPos[1] + hig] == '0':
          if [currentPos[0], currentPos[1] + hig] not in already:
            self.push([currentPos[0], currentPos[1] + hig])
            already.append([currentPos[0], currentPos[1] + hig])
        elif map[currentPos[0]][currentPos[1] + hig] == 'x':
            return [currentPos[0], currentPos[1] + hig]
      self.display()

      if not self.isEmpty():
        currentPos = self.pop()
      else:
        print("어디로도 갈수 없슴다.")
        return 0



map = [[ '1', '1', '1', '1', '1', '1'],
       ['e', '0', '0', '0', '0', '1'],
       ['1', '0', '1', '0', '1', '1'],
       ['1', '1', '1', '0', '0', 'x'],
       ['1', '1', '1', '0', '1', '1'],
       ['1', '1', '1', '1', '1', '1']]

search = Search()
print("출구:", search.deepSearch(map))
```
