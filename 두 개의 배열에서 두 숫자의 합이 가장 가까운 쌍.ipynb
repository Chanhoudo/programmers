```python
def find_closest_pair(A, B, target):
    # Step 1: 정렬
    A.sort()
    B.sort()

    # Step 2: 포인터 초기화
    left = 0
    right = len(B) - 1
    closest_sum = float('inf')
    closest_pair = None

    # Step 3: 투 포인터 방식으로 탐색
    while left < len(A) and right >= 0:
        current_sum = A[left] + B[right]

        # 타겟과의 차이를 확인
        if abs(current_sum - target) < abs(closest_sum - target):
            closest_sum = current_sum
            closest_pair = (A[left], B[right])

        # 합이 타겟보다 크다면 right 포인터를 줄이고, 작다면 left 포인터를 늘림
        if current_sum > target:
            right -= 1
        else:
            left += 1

    return closest_pair, closest_sum


# Test Example
A = [1, 3, 5, 7]
B = [10, 20, 30, 40]
target = 32

pair, closest_sum = find_closest_pair(A, B, target)
print(f"Closest pair: {pair}")
print(f"Closest sum: {closest_sum}")
```
